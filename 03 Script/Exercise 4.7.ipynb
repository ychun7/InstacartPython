```python
# define subset of ords_prods_merged
df = ords_prods_merged[:1000000]
```

```python
# define price labels
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'np.nan'
```

```python
#apply if statement to df
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# applying with loc statements
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# defining 2 busiest days and 2 slowest days 
new_result = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value == 0 or value == 1:
    new_result.append("Busiest days")
  elif value == 3 or value == 4:
    new_result.append("Slowest Days")
  else:
    new_result.append("Regularly busy")
```

```python
#create new column with new filter as Busiest_days
new ords_prods_merged['Busiest_days']=new_result
```

```python
#check count
ords_prods_merged['Busiest_days'].value_counts(dropna=False)
```

```python
#check frequency for order_hour_of_day
ords_prods_merged['order_hour_of_day'].value_counts(dropna=False)
```

```python
# create range for hours of day

hours_range = []

for value in ords_prods_merged["order_hour_of_day"]:
  if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16 or value == 9:
    hours_range.append("Most orders")
  elif value == 3 or value == 4 or value == 2 or value == 5 or value == 1 or value == 0 or value == 6 or value == 23:
    hours_range.append("Fewest orders")
  else:
    hours_range.append("Average orders")
```

```python
#create new column with new filter as busiest_period_of_day
ords_prods_merged['busiest_period_of_day']=hours_range
```

```python
#value count
ords_prods_merged['busiest_period_of_day'].value_counts(dropna=False)
```
