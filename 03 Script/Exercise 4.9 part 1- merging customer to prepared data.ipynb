```python
#define customer
df_cust= pd.read_csv(os.path.join(path,'02 Data','Original Data','customers.csv'), index_col = False)
```

```python
#define ords_prods_merge_with_flags 
ords_prods_merge=pd.read_pickle(os.path.join(path,'02 Data', 'Prepared Data','orders_products_merge_with_flags.pkl'))
```

```python
#rename Surname to Last Name
df_cust.rename(columns = {'Surnam':'Last Name'}, inplace=True)
```

```python
#rename STATE to State
df_cust.rename(columns = {'STATE':'State'}, inplace=True)
```

```python
#rename n_depedants to dependants
df_cust.rename(columns = {'n_dependants':'Dependents'},inplace=True)
```

```python
#rename fam_status to Family Status
df_cust.rename(columns = {'fam_status':'Family Status'},inplace=True)
```

```python
#check new headings
df_cust.head()
```

```python
#check for missing values
df_cust.isnull().sum()
```

```python
df_cust.shape
```

```python
# create new data frame without missing first names 
df_cust_new = df_cust[df_cust['First Name'].isnull()==False]
```

```python
#check count
df_cust_new.shape
```

```python
#check duplicates with new dataframe
df_dups = df_cust_new[df_cust_new.duplicated()]
```

```python
#check dups
df_dups
```

```python
#check for mixed type of data
for col in df_cust_new.columns.tolist(): weird = (df_cust_new[[col]].applymap(type) != df_cust_new[[col]].iloc[0].apply(type)).any(axis = 1)
if len (df_cust_new[weird]) > 0: print (col)
```

```python
#check data type for ords_prods_merge
ords_prods_merge.head()
```

```python
#check ords_prods_merge data types
ords_prods_merge.dtypes
```

```python
#check ords_prods_merge shape
ords_prods_merge.shape
```

```python
#combine df_cust_new with ords_prods_merge on user_id
df_merge = ords_prods_merge.merge(df_cust_new, on = 'user_id', indicator = 'merge')
```
