```python
# grouping by product name - step 1 splitting data
df.groupby('product_name')
```

```python
#calculate mean of order number grouped by department_id
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# calculate mean of order number grouped by department id without agg function
df.groupby('department_id')['order_number'].mean()
```

```python
#calculate mean, min, and max of order numbers
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
#create columns - maximum number of order for each user 
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
#create flag for loyalty customers based on max order
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#check count
ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
#check basic stastics of loyalty flags - loyal customers
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

The above statistics show that mean of prices purchased by loyal customers are not the highest. The price mean is actually highest for new customers and even regular customers have higher price mean compared to loyal customers.

```python
#create column for price average
ords_prods_merge['avg_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#check head
ords_prods_merge.head()
```

```python
#create flag for spending based on average price - high spender 
ords_prods_merge.loc[ords_prods_merge['avg_price'] >=10, 'spending_flag'] = 'high spender'
```

```python
#create flag for spending based on avg price - low spender
ords_prods_merge.loc[ords_prods_merge['avg_price'] <10, 'spending_flag'] = 'low spender'
```

```python
#check count
ords_prods_merge['spending_flag'].value_counts(dropna=False)
```

```python
#create column for median of days_since_prior_order
ords_prods_merge['days_since_prior_median'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#check head
ords_prods_merge.head()
```

```python
#create flag for frequency based on days_since_prior_order median - frequent customer
ords_prods_merge.loc[ords_prods_merge['days_since_prior_median'] <=10, 'frequency_flag'] = 'frequent customer'
```

```python
#create flag for frequency based on days_since_prior_order median - non-frequent customer
ords_prods_merge.loc[ords_prods_merge['days_since_prior_median'] >20, 'frequency_flag'] = 'non-frequent customer'
```

```python
#create flag for frequency based on days_since_prior_order median - regular customer
ords_prods_merge.loc[(ords_prods_merge['days_since_prior_median'] >10) & (ords_prods_merge['days_since_prior_median']<=20), 'frequency_flag'] = 'regular customer'
```

```python
#check count
ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
